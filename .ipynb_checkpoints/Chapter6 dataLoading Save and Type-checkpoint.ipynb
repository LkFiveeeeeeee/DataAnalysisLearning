```python
with open('examples/ex7.csv') as f:
    lines = list(csv.reader(f))
```

```python
lines
```

```python
header
```

```python
values
```

```python
data_dict = {h: v for h, v in zip(header,zip(*values))}

data_dict
```

```python
f = open('examples/ex7.csv')
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL
reader = csv.reader(f, dialect = my_dialect)
```

```python
reader
```

```python
for line in reader:
    print(line)
```

```python
# pickle仅建议用于短期存储格式。其原因是很难保证该格式永远是稳定的。
# 今天的pickle对象可能无法被后续版本的库unpickle出来
```

```python
# pandas内置支持两个二进制数据格式  HDF5、MessagePack
```

```python
# HDF5是一种存储大规模科学数组数据的非常好的文件格式
# 它可以作为C标准库,带有许多语言的借口,如Java,Python和MaltLab
```

```python
import numpy as np
frame = pd.DataFrame({'a':np.random.randn(100)})
```

```python
store = pd.HDFStore('mydata.h5')

store['obj1'] = frame

store['obj1_col'] = frame['a']
```

```python
# Web APIs 交互
import requests
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
resp = requests.get(url)
```

```python
resp
```

```python
# 响应对象的json方法会返回一个包含解析过的JSON字典,加载到一个Python对象中

data = resp.json()
```

```python
data
```

```python
data[0]['title']
```

```python
issues = pd.DataFrame(data,columns=['number','title',
                                   'labels','state','user'])
```

```python
issues
```
