```python
# 使用DataFrame的列进行索引
```

```python
# 人们经常想要将DataFrame的一个或多个列当做行索引来用
# 或者可能希望将行索引变成DataFrame的列

frame = pd.DataFrame({'a': range(7), 'b': range(7,0,-1),
                     'c':['one','one','one','two','two','two','two'],
                     'd':[0,1,2,0,1,2,3]})
```

```python
frame
```

```python
# DataFrame的set_index函数会将其一个或多个列转换为行索引
frame2 = frame.set_index(['c','d'])
```

```python
frame2
```

```python
# 默认情况下,那些列会从DataFrame中移除,但有可以保留下来
frame.set_index(['c','d'],drop=False)
```

```python
# reset_index的功能与set_index相反
# 层次化索引的级别会被转移到列里面
frame2.reset_index()
```

```python
# 合并数据集
```

```python
## 数据库风格的DataFrame合并

## 数据库的合并或连接运算是通过一个或多个键将行连接起来的
## 这些运算时关系型数据库(基于SQL)的核心
## pandas的merge函数是对数据应用这些算法的主要切入点
```

```python
df1 = pd.DataFrame({'key':['b','b','a','c','a','a','b'],
                   'data1':range(7)})

df2 = pd.DataFrame({'key':['a','b','d'],
                   'data2':range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
# 如果两个对象的列名不同,也可以分别进行指定
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                   'data1': range(7)})

df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                   'data2': range(3)})
```

```python
pd.merge(df3,df4,left_on='lkey',right_on='rkey')
```

```python
# 默认情况下,merge做的是"内连接" 结果中的键是交集
# 也可取左连接、右连接和外连接

pd.merge(df1,df2,how='outer')
```

```python
# 索引上的合并
```

```python
# 有时候,DataFrame中的连接键位于其索引中
# 这种情况下,可以传入left_index=True或right_index=True以说明索引应该被用作连接键
```

```python
left1 = pd.DataFrame({'key':['a','b','a','a','b','c'],
                     'value':range(6)})
```

```python
right1 = pd.DataFrame({'group_val':[3.5,7]},index=['a','b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1,right1,left_on='key',right_index=True)
```

```python
# 层次化索引合并

lefth = pd.DataFrame({'key1':['Ohio','Ohio','Ohio',
                             'Nevada','Nevada'],
                     'key2':[2000,2001,2002,2001,2002],
                     'data':np.arange(5)})
```

```python
lefth
```

```python
righth = pd.DataFrame(np.arange(12).reshape((6,2)),
                     index=[['Nevada','Nevada','Ohio','Ohio'
                            ,'Ohio','Ohio'],
                           [2001,2000,2000,2000,2001,2002]],
                     columns=['event1','event2'])
```

```python
righth
```

```python
# 这种情况下必须以列表的形式指明用作合并键的多个列

pd.merge(lefth,righth,left_on=['key1','key2'],right_index=True)
```

```python
# 还可以同时合并双方索引
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
   ....:                      index=['a', 'c', 'e'],
   ....:                      columns=['Ohio', 'Nevada'])
```

```python
 right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
   ....:                       index=['b', 'c', 'd', 'e'],
   ....:                       columns=['Missouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2,right2,how='outer',left_index=True,right_index=True)
```

```python
# DataFrame还有一个便捷的join实例方法
# 可以更为方便的实现按索引合并
# 还可以用于合并多个带有相同或相似索引的DataFrame对象
# !! 但要求没有重叠的列

left2.join(right2,how='outer')
```

```python
# 重塑和轴向旋转
```

```python
# 重塑层次化索引
# 层次化索引为DataFrame数据的重排任务提供了一种具有良好一致性的方式
# 主要功能有二
# stack: 将数据的列 旋转 为行
# unstack: 将数据的行 旋转 为列
```

```python
data = pd.DataFrame(np.arange(6).reshape((2,3)),
                   index=pd.Index(['Ohio','Colorado'],name='state'),
                   columns=pd.Index(['one','two','three'],name='number')
                   )
```

```python
data
```

```python
# 对该数据使用stack方法即可将列转换为行,得到一个Series
result = data.stack()
```

```python
result
```

```python
# 对于一个层次化索引的Series
# 可以使用unstack将其重排位一个DataFrame
result.unstack()
```

```python
# 默认情况下,unstack操作的是最内层。
# 传入分层级别的编号或名称即可对其他级别进行操作

result.unstack(0)
```

```python
# 等同于
result.unstack('state')
```

```python
# 对DataFrame进行unstack操作时,作为旋转轴的级别将会成为结果中的最低级别

df = pd.DataFrame({'left': result, 'right': result + 5},
                 columns=pd.Index(['left', 'right'], name='side'))
```

```python
df
```

```python
df.unstack()
```

```python
data = pd.read_csv('examples/macrodata.csv')
```

```python
periods = pd.PeriodIndex(year=data.year,quarter=data.quarter,
                        name='data')
```

```python
periods
```

```python
columns = pd.Index(['realgdp','infl','unemp'],name='item')
```

```python
columns
```

```python
data = data.reindex(columns=columns)
```

```python
data
```

```python
data.index = periods.to_timestamp('D','end')
```

```python
data
```

```python
ldata = data.stack().reset_index().rename(columns={0:'value'})
```

```python
ldata
```

```python
pivoted = ldata.pivot('data','item','value')
```

```python
pivoted
```

```python
# 将"宽格式"旋转为"长格式"
# 旋转DataFrame的逆运算时pandas.melt
# 它不是将一列转换到多个新的DataFrame,而是合并多个列成为一个长的DataFrame
```

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                  'A': [1, 2, 3],
                  'B': [4, 5, 6],
                  'C': [7, 8, 9]})
```

```python
df
```

```python
# key列可能是分组指标，其它的列是数据值。
# 当使用pandas.melt，我们必须指明哪些列是分组指标
melted = pd.melt(df,['key'])
```

```python
melted
```

```python
# 使用pivot,可以重塑回原来的样子

reshaped = melted.pivot('key','variable','value')
```

```python
reshaped
```

```python
# pivot的结果从列中创建了一个索引,用作行标签
# 我们可以使用reset_index将数据移回列
reshaped.reset_index()
```
