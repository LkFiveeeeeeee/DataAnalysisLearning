```python
# 调整subplot周围的间距
# 默认情况下,matplotlib会在subplot外围留下一定的边距
# 间距跟图像的高度和宽度有关
# 利用Figure的subplots_adjust方法可以轻而易举地修改间距

# subplot_adjust(...)
```

```python
# wspace和hspace用于控制宽度和高度的百分比

fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50,color='k',alpha = 0.5)
        
plt.subplots_adjust(wspace = 0,hspace=0)
```

```python
# 注解和文字可以通过text、arrow和annotate函数进行添加
```

```python
from datetime import datetime
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
```

```python
import pandas as pd
data = pd.read_csv('examples/spx.csv',index_col=0,parse_dates=True)
spx = data['SPX']
```

```python
spx.plot(ax=ax,style='k-')
```

```python
crisis_data=[(datetime(2007,10,11),'Peak of bull market'),
            (datetime(2008,3,12),'Bear Stearns Fails'),
            (datetime(2008,9,15),'Lehman Bankruptcy')]
```

```python
for date,label in crisis_data:
    ax.annotate(label,xy=(date,spx.asof(date) + 75),
               xytext=(date,spx.asof(date) + 225),
               arrowprops=dict(facecolor='black',headwidth=4,width=2,
                              headlength=4),
               horizontalalignment='left',verticalalignment='top')
```

```python
ax.set_xlim(['1/1/2007','1/1/2011'])
```

```python
ax.set_ylim([600,1800])
```

```python
ax.set_title("Important dates in the 2008-2009 financial crisis")
```

```python
# 线型图
# Series和DataFrame都有一个用于生成各类图表的plot方法
# 默认情况下,生成的是线型图
```

```python
s = pd.Series(randn(10).cumsum(),index=np.arange(0,100,10))
s.plot()
```

```python
# DataFrame的plot方法会在一个subplot中为各列绘制一条线,并自动创建图例
df = pd.DataFrame(np.random.randn(10,4).cumsum(0),
                 columns=['A','B','C','D'],
                 index=np.arange(0,100,10))
```

```python
df.plot()
```

```python
# 柱状图
# plot.bar() 和plot.barh()分别绘制垂直和水平的柱状图 Series和DataFrame的索引会被用作刻度
```

```python
fig,axes = plt.subplots(2,1)
data = pd.Series(np.random.rand(16),index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0],color='k',alpha=0.7)
```

```python
data.plot.barh(ax=axes[1],color='k',alpha=0.7)
```

```python
# 对于DataFrame 柱状图会将每一行的值分为一组,并排显示
df = pd.DataFrame(np.random.rand(6,4)
                 ,index=['one','two','three','four','five','six'],
                 columns=pd.Index(['A','B','C','D'], name='Genus'))
```

```python
df
```

```python
df.plot.bar()
```

```python
# 设置stacked=True即可为DataFrame生成堆积柱状图，
df.plot.barh(stacked=True,alpha=0.4)
```

```python
# 假设我们想要做一张堆积柱状图以展示每天各种聚会规模的数据点的百分比
```

```python
tips = pd.read_csv('examples/tips.csv')
```

```python
tips.head()
```

```python
party_counts = pd.crosstab(tips['day'],tips['size'])
```

```python
party_counts
```

```python
party_counts = party_counts.loc[:, 2 : 5]
```

```python
party_counts
```

```python
party_pcts = party_counts.div(party_counts.sum(1),axis=0)
```

```python
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
# 对于在绘制一个图形之前,需要合计的数据,使用seaborn可以减少工作量
```

```python
import seaborn as sns
```

```python
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```python
sns.barplot(x='tip_pct',y='day',data=tips,orient='h')
# 黑线代表95%的置信区间
```

```python
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
```

```python
# 散布图或点图
# 点图或散布图是观察两个一维数据序列之间的关系的有效手段

macro = pd.read_csv('examples/macrodata.csv')
```
