```python
import pandas as pd
import numpy as np
```

```python
# 数据聚合
# 聚合指的是任何能够从数组产生标量值的数据转换过程
```

```python
# 如果需要使用自己的聚合函数,则需将其传入aggregate或agg方法

def peak_to_peak(arr):
    return arr.max() - arr.min()

grouped.agg(peak_to_peak)
```

```python
# 自定义聚合函数要比经过优化的函数慢得多
# 因为在构造中间分组数据块时存在非常大的开销
```

```python
# 面向列的多函数应用
```

```python
tips = pd.read_csv('examples/tips.csv')
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
tips[:6]
```

```python
# 人们可能希望对不同的列使用不同的聚合函数
# 或一次应用多个函数
```

```python
# exp
# 根据day和smoker对tips进行分组

grouped = tips.groupby(['day','smoker'])
```

```python
grouped
```

```python
grouped_pct = grouped['tip_pct']
```

```python
grouped_pct
```

```python
grouped_pct.agg('mean')
# 可以将函数名以字符串的形式传入
```

```python
# 如果传入一组函数或函数名
# 得到的DataFrame的列就以相应的函数命名

grouped_pct.agg(['mean','std',peak_to_peak])
```

```python
# 假设你要对一个列或不同的列应用不同的函数
# 方法是向agg传入一个从列名映射到函数的字典

grouped.agg({'tip':np.max,'size':'sum'})
```

```python
grouped.agg({'tip_pct' : ['min', 'max', 'mean', 'std'],
   ....:              'size' : 'sum'})
```

```python
# 行索引不是必要的,可以通过as_index来去掉
tips.groupby(['day', 'smoker'], as_index=False).mean()
```

```python
def top(df,n=5,column='tip_pct'):
    return df.sort_values(by=column)[-n:]

top(tips,n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
# 如果传给apply的函数能够接受其他参数或关键字
# 则可以将这些内容放在函数名后面一并参入
tips.groupby(['smoker','day']).apply(top
                               ,n=1 ,column='total_bill')
```

```python
tips.groupby(['smoker','day']).apply(top
                                ,column='total_bill')
```

```python
# 禁止分组键
# 从上面的例子中可以看出
# 分组键会跟原始对象的索引共同构成结果对象中的层次化索引
#将group_keys=False传入groupby即可禁止该效果

tips.groupby('smoker',group_keys=False).apply(top)
```

```python
# 分位数和桶分析
frame = pd.DataFrame({'data1': np.random.randn(1000),
                     'data2':np.random.randn(1000)})

quartiles = pd.cut(frame.data1,4)
```

```python
quartiles
```

```python
quartiles[:10]
```

```python
def get_stats(group):
    return {'min':group.min(),
           'max':group.max(),
           'count':group.count(),
           'mean':group.mean()}

grouped = frame.data2.groupby(quartiles)

grouped.apply(get_stats)
```

```python
grouped.apply(get_stats).unstack()
```

```python
grouping = pd.qcut(frame.data1,10,labels = False)

grouped = frame.data2.groupby(grouping)

grouped.apply(get_stats).unstack()
```

```python
# 示例: 用特定于分组的值填充缺失值

s = pd.Series(np.random.randn(6))

s[::2] = np.nan

s
```

```python
s.fillna(s.mean())
```

```python
# 假设你需要对不同的分组填充不同的值
# 一种方法是将数据分组
# 使用apply和一个能够对各数据块调用fillna的函数即可

states =  {'Ohio','New York',
          'Vermont','Florida',
          'Oregon','Nevada','California',
          'Idaho'}

group_key = ['East'] * 4 + ['West'] * 4
group_key
```

```python
data = pd.Series(np.random.randn(8),
                index=states)
```

```python
data
```

```python
data[['Vermont','Nevada','Idaho']] = np.nan
```

```python
data
```

```python
data.groupby(group_key).mean()
```

```python
# 我们可以用分组平均值去填充NA值
fill_mean = lambda g: g.fillna(g.mean())

data.groupby(group_key).apply(fill_mean)
```

```python
fill_values = {'East':0.5,'West':-1}
fill_func = lambda g:g.fillna(fill_values[g.name])

data.groupby(group_key).apply(fill_func)
```

```python
# 随机采样和排列
# 假设你想要从一个大数据集中随机抽取
# (进行替换或不替换)
# 样本以进行蒙特卡罗模拟或其他分析工作

# 这里采用对Series使用sample方法
```

```python
suits = ['H','S','C','D']
```

```python
card_val = (list(range(1,11)) + [10] * 3) * 4 
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
```

```python
card_val
```

```python
base_names
```

```python
cards=[]
```

```python
for suit in ['H','S','C','D']:
    cards.extend(str(num) + suit for num in base_names)
```

```python
cards
```

```python
deck = pd.Series(card_val,index=cards)
```

```python
deck
```

```python
# 从中抽取5张
def draw(deck,n=5):
    return deck.sample(n)

draw(deck)
```

```python
get_suit = lambda x:x[-1]
deck.groupby(get_suit).apply(draw,n=2)
```

```python
deck.groupby(get_suit,group_keys=False).apply(draw,n=2)
```

```python
# 分组加权平均数和相关系数
# 根据groupby的"拆分-应用-合并"范式
# 可以进行DataFrame的列与列之间或两个Series之间的运算

df = pd.DataFrame({
    'category':['a','a','a','a',
               'b','b','b','b'],
    'data':np.random.randn(8),
    'weights':np.random.rand(8)
})

df
```

```python
grouped = df.groupby('category')
```

```python
get_wavg = lambda g:np.average(
    g['data'],weights=g['weights']
    )
grouped.apply(get_wavg)
```

```python
close_px = pd.read_csv('examples/stock_px_2.csv',
                      parse_dates=True,
                      index_col=0)
close_px.info()
```

```python
close_px[-4:]
```

```python
# 计算一个由日收益率(通过百分数变化计算)与
# SPX之间的年度相关系数组成的DataFrame

spx_corr = lambda x:x.corrwith(x['SPX'])

rets = close_px.pct_change().dropna()
```

```python
get_year = lambda x: x.year
by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
by_year.apply(lambda g:g['AAPL'].corr(g['MSFT']))
```

```python
# 组级别的线性回归

import statsmodels.api as sm
def regress(data,yvar,xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1
    result = sm.OLS(Y,X).fit()
    return result.params
```

```python
by_year.apply(regress,'AAPL',['SPX'])
```
